```python
# import data
# try 1 data set first
import pandas as pd
# can change the file name (refer to files folder)
df = pd.read_excel("./files/2022-23_T1.xlsx")
```

```python
# check head
print(df.head())
print("")
# check if columns are there
for column in df.columns:
    print("Column Name:",column)
```

```python
# throw "Term" and "Session" as they are likely to be repeated
df = df.rename(columns={"Bidding Window":"Window","Course Code":"Code","Description":"ModName",
                        "Opening Vacancy":"Opening Vacancy",
               "Before Process Vacancy":"BProcessVacancy","D.I.C.E":"DICE",
                "After Process Vacancy":"AProcessVacancy","Enrolled Students":"Enrolled",
               "Median Bid":"Median","Min Bid":"Min","Instructor":"Prof",
               "School/Department":"School"}) 
df = df.drop(['Term','Session'],axis=1)
# replace column name
print(df.head())

```

```python
# get all IS mods price in round 1 window 1
# allISMods = df.query(["School == 'SIS'","Window == 'Round 1 Window 1'"])
# allISMods.head()
allISMods = df.loc[(df['School'] == 'SIS') & (df['Window'] == 'Round 1 Window 1'),["Code","ModName","Prof","Median","Min"]]
allISMods.head()
```

```python
# group by 
groupedMods = allISMods.groupby(["Code","ModName"])['Median','Min'].mean()
# .agg(["Median","Min"])
groupedMods.round(2).sort_values(by='Median', ascending=True)
```
